````python
a = float(input("Введите длину первого отрезка a: "))
b = float(input("Введите длину второго отрезка b: "))
c = float(input("Введите длину третьего отрезка c: "))

# Проверка неравенства треугольника
if a + b > c and a + c > b and b + c > a:
    is_triangle = True
else:
    is_triangle = False

# Проверка на прямоугольность (теорема Пифагора)
if is_triangle:
    sides = [a, b, c]
    sides.sort()
    if sides[2] ** 2 == sides[0] ** 2 + sides[1] ** 2:
        is_right_triangle = True
    else:
        is_right_triangle = False
else:
    is_right_triangle = False

if is_triangle:
    if is_right_triangle:
        print("Можно составить треугольник, и это прямоугольный треугольник.")
    else:
        print("Можно составить треугольник, но это не прямоугольный треугольник.")
else:
    print("Нельзя составить треугольник.")
````

8. Напишите программу для решения уравнения
$$ ax^2+bx+c=0 $$
Коэффициенты $a, b, c$ могут быть любыми числами. 

````python
import math

a = float(input("Введите коэффициент a: "))
b = float(input("Введите коэффициент b: "))
c = float(input("Введите коэффициент c: "))

# Вычисляем дискриминант
discriminant = b**2 - 4*a*c

# Проверяем значения дискриминанта и находим корни
if discriminant > 0:
    x1 = (-b + math.sqrt(discriminant)) / (2*a)
    x2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"У уравнения два корня: x1 = {x1}, x2 = {x2}")
elif discriminant == 0:
    x1 = -b / (2*a)
    print(f"У уравнения один корень: x1 = {x1}")
else:
    print("Уравнение не имеет действительных корней")
````

11. Выведите значение заданного целого числа от 0 до 999 прописью. Например, «сто девяносто один» для числа 191, «одиннадцать» для числа 11. 

````python
def number_to_words(number):
    units = ["", "один", "два", "три", "четыре", "пять", "шесть", "семь", "восемь", "девять"]
    teens = ["", "одиннадцать", "двенадцать", "тринадцать", "четырнадцать", "пятнадцать", "шестнадцать", "семнадцать", "восемнадцать", "девятнадцать"]
    tens = ["", "десять", "двадцать", "тридцать", "сорок", "пятьдесят", "шестьдесят", "семьдесят", "восемьдесят", "девяносто"]
    hundreds = ["", "сто", "двести", "триста", "четыреста", "пятьсот", "шестьсот", "семьсот", "восемьсот", "девятьсот"]

    if 0 <= number <= 999:
        if number == 0:
            return "ноль"
        else:
            hundreds_digit = number // 100
            tens_digit = (number % 100) // 10
            units_digit = number % 10
            result = hundreds[hundreds_digit]
            if tens_digit == 1:
                result += " " + teens[units_digit]
            else:
                result += " " + tens[tens_digit]
                if units_digit != 0:
                    result += " " + units[units_digit]
            return result

number = int(input("Введите целое число от 0 до 999: "))

if 0 <= number <= 999:
    number_in_words = number_to_words(number)
    print(number_in_words)
else:
    print("Число не входит в диапазон от 0 до 999")
````

Реализовать калькулятор, который принимает от пользователя через пробел строку следующего вида: «a op b», где a и b – некоторые числа, а ‘op’ определяет оператор и может принимать значения $«+,-,*,/,**,%»$. В зависимости от оператора с помощью форматирования строк вывести результат в виде: «a + b = 3 + 2 = 5», где была получена строка «3 + 2». Для операций возведения в степень и деления по модулю использовать вместо знака оператора соответствующие выражения. Например, для строки «2 ** 3» должно быть выведено «a в степени b = 2 в степени 3 = 8».

````python
expression = input("Введите выражение в формате 'a op b' (например, '3 + 2'): ")
a, op, b = expression.split()

a = float(a)
b = float(b)

if op == '+':
    result = a + b
    op_str = "+"
elif op == '-':
    result = a - b
    op_str = "-"
elif op == '*':
    result = a * b
    op_str = "*"
elif op == '/':
    result = a / b
    op_str = "/"
elif op == '**':
    result = a ** b
    op_str = "в степени"
else:
    print("Неверный оператор. Поддерживаются операторы: +, -, *, /, **")
    exit()

formatted_expression = f"{a} {op_str} {b} = {a} {op_str} {b} = {result}"

print(formatted_expression)

````

```
 (~~0~~) (~~0~~) (~~0~~)
```

16. Имеются два n – мерных вектора x и y, которые задают координаты n точек на плоскости (случайные целые числа). Найти наиболее близкие друг другу точки.

````python
import random
import math

def generate_random_vector(n):
    return [random.randint(1, 100) for _ in range(n)]

def calculate_distance(point1, point2):
    return math.sqrt(sum((p1 - p2) ** 2 for p1, p2 in zip(point1, point2)))

n = int(input("Введите размерность векторов (n): "))
num_points = int(input("Введите количество точек: "))

x = [generate_random_vector(n) for _ in range(num_points)]
y = [generate_random_vector(n) for _ in range(num_points)]

min_distance = float("inf")
closest_points = None

for i in range(num_points):
    for j in range(num_points):
        if i != j:
            distance = calculate_distance(x[i], y[j])
            if distance < min_distance:
                min_distance = distance
                closest_points = (x[i], y[j])

print("Наиболее близкие точки:")
print("x:", closest_points[0])
print("y:", closest_points[1])
print("Расстояние между ними:", min_distance)
````

17. Из множества целых чисел от 1 до N выделить множество N2 числа, кратные 2, множество N3 – кратные 3, множество N 6 – кратные 6 (т.е. кратные и 2 и 3), множество N 23 – кратные либо 2, либо 3.

````python
N = int(input("Введите значение N: "))

# Множество чисел, кратных 2
N2 = set(i for i in range(1, N + 1) if i % 2 == 0)

# Множество чисел, кратных 3
N3 = set(i for i in range(1, N + 1) if i % 3 == 0)

# Множество чисел, кратных 6 (кратных и 2 и 3)
N6 = N2.intersection(N3)

# Множество чисел, кратных либо 2, либо 3
N23 = N2.union(N3)

print("Множество N2 (кратные 2):", N2)
print("Множество N3 (кратные 3):", N3)
print("Множество N6 (кратные 6):", N6)
print("Множество N23 (кратные 2 или 3):", N23)
````

````python
# Введите слова, разделенные пробелами
words = input("Введите слова, разделенные пробелами: ")

# Введите букву, количество которой нужно найти
letter = input("Введите букву, количество которой нужно найти: ")

# Разделите введенные слова на отдельные слова
word_list = words.split()

# Инициализируйте счетчик
count = 0

# Пройдитесь по каждому слову и посчитайте буквы
for word in word_list:
    count += word.count(letter)

print(f"Общее количество букв '{letter}' в ваших словах: {count}")
````

19. Даны m (m>1) слов. В каком из них доля (в %) заданной буквы больше.

````python
# Введите слова, разделенные пробелами
words = input("Введите слова, разделенные пробелами: ")

# Введите букву, долю которой нужно найти
letter = input("Введите букву, долю которой нужно найти: ")

# Разделите введенные слова на отдельные слова
word_list = words.split()

# Инициализируйте переменные для отслеживания наибольшей доли и соответствующего слова
max_percentage = 0
word_with_max_percentage = ""

# Пройдитесь по каждому слову и найдите долю заданной буквы
for word in word_list:
    total_characters = len(word)
    letter_count = word.count(letter)
    percentage = (letter_count / total_characters) * 100

    if percentage > max_percentage:
        max_percentage = percentage
        word_with_max_percentage = word

print(f"В слове '{word_with_max_percentage}' доля буквы '{letter}' составляет {max_percentage:.2f}% и больше, чем в других словах.")
````
